## LLM: Sentiment Analysis

```python
# create prompt to ask the language model
def create_prompt(text):
  """Create prompt ot the language model.
  Args:
    text: prompt for text classification
  return:
    prompt for text classification
    """
  instruction = 'Is the underlying sentiment positive or negative'
  formatting = '"Positive" or "Negative"' # how you want your responce to be
  return f'Text: {text}\n {instruction}\n Answer {formatting}'
```

```python
# the function call the language model using the prompt
def call_llm(prompt):
  """Call the LLM with the input prompt and return the answer.
  Arg:
    prompt: Prompt to sent to the LLM
  return:
    Answer: Answer generated by the LLM
    """
  message= [{'role':'user',
              'content':prompt}]  # Here we add the role (user/assistant) and the content of the message
  response= client.chat.completions.create(messages=message,
                                            model='gpt-4o')
  return response.choices[0].message.content



```

```python
# Processing the csv file we add another function to classify
def classify(text):
  """Classify input text.
  Arg:
    text: Text to classify
  return
    class label eg. Positive or Negative
  """
  prompt=create_prompt(text)
  answer= call_llm(prompt)
  return answer

# argument parsing
parser=argparse.ArgumentParser()
parser.add_argument('path',type=str,help='path to the csv file')
args=parser.parse_args()
df=pandas.read_csv(args.path)
df['class']=df['text'].apply(classify)
```

## MULTI-ODEL LLM:Image input

```python
def analyze_image(image_url,question):
  """ Answer the question about the image."""
  message=[{'role':'user','content':[
      {'type':'text','text':question},
      {'type':'image_url','image_url':{'url':'image_url'}}
  ]}]
  reply=client.chat.competions.create(
      model='gpt-4'
      messages=message
  )
  return reply.choice[0].message.content

parser= argparse.ArgumentParser()
parser.add_argument('image_url',typee=str,help="image url")
parser.add_argument('question',type=str,help='question about the image')
args=parser.parse_args()
answer=analyze_image(args.image_url,args.question)
```

```python
def create_prompt(text):
  """Create input prompt for language model.
  Arg:
    text: text to classify
  Return:
    prompt for text classification
  """
  instruction='Is the underlying sentiment positive or negative?'
  formatting = '"Positive" or "Negative"'
  return f'Text:{text}\n{instruction}\n answer: {formatting}'
```

```python
def call_llm(prompt):
    """Call LLaMA with the input prompt and return the answer.
    Args:
        prompt: Prompt to send to the language model
    Returns:
        Answer generated by LLaMA model
    """
    # Load LLaMA tokenizer and model
    tokenizer = LlamaTokenizer.from_pretrained("meta-llama/Llama-2-70b-hf")
    model = LlamaForCausalLM.from_pretrained("meta-llama/Llama-2-70b-hf")

    # Tokenize the prompt and generate response
    inputs = tokenizer(prompt, return_tensors="pt")

    # Make sure to move the model to the appropriate device (CPU or GPU)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Generate output from the model
    with torch.no_grad():
        outputs = model.generate(inputs.input_ids.to(device), max_length=50)

    # Decode the output
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```
